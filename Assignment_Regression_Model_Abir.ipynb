```python
#separating input and output
x = d_set[['lstat','indus','nox','ptratio','rm','tax','dis','age']]   #feature
y = d_set['medv'] #target
x.head()
```

```python
y.head()
```

# Creating Model

```python
#creating model for train and test sets and splitting
from sklearn.model_selection import train_test_split
X_train , X_test ,Y_train , Y_test = train_test_split(x,y,test_size =.3, random_state=45)
```

```python
print(X_train.shape , X_test.shape ,Y_train.shape , Y_test.shape)
```

# Training the Model

```python
#importing Liner Regression Class
from sklearn.linear_model import LinearRegression
```

```python
l_regression = LinearRegression()
```

```python
#fitting the train data
l_regression.fit(X_train,Y_train)
```

```python
l_regression.intercept_
```

```python
l_regression.coef_
```

```python
co_df = pd.DataFrame(data = l_regression.coef_, index = x.columns, columns = ['Coefficient'])
co_df
```

```python
#checking the prediction from our testset
# So if we increase room per dwelling then the owner occupied homis median value will also be increase
prediction = l_regression.predict(X_test)
print(prediction[:20])
```
